```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
df = pd.read_excel('Iris.xls')

df_labels = df['iris']
labels = df_labels.values

# convert labels to numbers
labels_num = []

for i in labels:
    if i == 'Iris-setosa':
        labels_num.append(1)
    elif i == 'Iris-versicolor':
        labels_num.append(2)
    else:
        labels_num.append(3)

df_labels_num = pd.DataFrame(labels_num)

df['iris'] = df_labels_num 
```

```python
index = [i for i in range(df.shape[0])]
random.shuffle(index)
df = df.set_index([index]).sort_index()
df = df.iloc[:-10]
df_unknown = df.iloc[-10:]
```

```python
df
```

```python
class KNN():
    def __init__(self,df,df_unknown):
        self.df = df.values
        self.df_unknown = df_unknown.values
        self.top_3_distances_and_classes = []
        
    def get_distances(self):
        distances_and_classes = []
        
        for i,data_df_unknown in enumerate(self.df_unknown):
            a = []

            for j,data_df in enumerate(self.df):
                euc_dis = np.sqrt((data_df[0] - data_df_unknown[0])**2 + 
                                  (data_df[1] - data_df_unknown[1])**2 + 
                                  (data_df[2] - data_df_unknown[2])**2 + 
                                  (data_df[3] - data_df_unknown[3])**2)

                a.append([round(euc_dis,2), data_df[4]])
            distances_and_classes.append(a)
            
        sorted_distances_and_classes = [] 
    
        for i in distances_and_classes:
            new_list = sorted(i, key=lambda x: x[0])
            sorted_distances_and_classes.append(new_list)

        for i in sorted_distances_and_classes:
            top_3 = []
    
            for j in range(3):
                top_3.append(i[j])
        
            self.top_3_distances_and_classes.append(top_3)
        return self.top_3_distances_and_classes

    def get_accuracy(self):
        pred_list = []
        actual = []
        for i in self.df_unknown:
            actual.append(i[-1])
            
        for i in self.top_3_distances_and_classes:
            votes = []
            
            for j in i:
                votes.append(j[-1])
            pred_list.append(votes)
        
        pred = []
        
        for i in pred_list:
            holder = max(set(i), key=i.count)
            pred.append(holder)
         
        count = 0
        for pre,act in zip(pred, actual):
            if pre == act:
                count +=1
                
        return 'accuracy: {}/{}'.format(count,len(pred))
```

```python
iris_KNN = KNN(df,df_unknown)
```

```python
iris_KNN.get_distances()
# 3 (distance, class num) for each 
```

```python
iris_KNN.get_accuracy()
```
